# Analyse values in emp_length and revol_util and replace NA values with suitable changes

```python
df_Impact.emp_length.mode()
```

Replacing the NA values with mode as the impact will be minimal because the % to be replaced is less tha 3%

```python
print(df_Impact.emp_length.isna().sum())
df_Impact.emp_length.fillna(df_Impact.emp_length.mode()[0],inplace=True)
print(df_Impact.emp_length.isna().sum())
```

Replacing %,k to carry out calculation for int rate,inc group,inc

```python
df_Impact.revol_util=df_Impact.revol_util.apply(lambda x: str(x).replace('%',''))
df_Impact.revol_util=df_Impact.revol_util.astype(float)
df_Impact.int_rate=df_Impact.int_rate.apply(lambda x: str(x).replace('%',''))
df_Impact.int_rate=df_Impact.int_rate.astype(float)
df_Impact.funded_amnt_inv=df_Impact.funded_amnt_inv.astype(int)
df_Impact.annual_inc=df_Impact.annual_inc.astype(float)
(df_Impact.isna().sum()/len(df_Impact.index))*100
df_Impact.shape
df_Impact.head(50)
```

# Creating Bins for 

```python
df_Impact['Int_rate_Rng'] = pd.cut(df_Impact['int_rate'], bins=5,precision =0,labels=['5-9','9-13','13-17','17-21','21-24'])
df_Impact['Open_acc_Rng'] = pd.cut(df_Impact['open_acc'],bins = 5,precision =0,labels=['2-10','10-19','19-27','27-36','36-44'])
df_Impact['Revol_util_Rng'] = pd.cut(df_Impact['revol_util'], bins=5,precision =0,labels=['0-20','20-40','40-60','60-80','80-100'])
df_Impact['Total_acc_Rng'] = pd.cut(df_Impact['total_acc'], bins=5,precision =0,labels=['2-20','20-37','37-55','55-74','74-90'])
df_Impact['Annual_Inc_Rng'] = pd.cut(df_Impact['annual_inc'], bins = [3000, 31000, 58000, 85000,112000,140000],precision =0,labels =['0k-31k','31k-58k','58k-85k','85k-112k','112k-140k'])
df_Impact['Installment_Rng'] = pd.cut(df_Impact['installment'], bins=10,precision =0,labels=['14-145','145-274','274-403','403-531','531-660','660-789','789-918','918-1047','1047-1176','1176-1305'])
df_Impact['Funded_amnt_inv_Rng'] = pd.cut(df_Impact['funded_amnt_inv'], bins=7,labels=['0-5k','5k-10k','10k-15k','15k-20k','20k-25k','25k-30k','30k-35k']) 
df_Impact['Loan_amnt_Rng'] = pd.cut(df_Impact['loan_amnt'], bins=7,precision =0,labels=['0-5k','5k-10k','10k-15k','15k-20k','20k-25k','25k-30k','30k-35k'])
df_Impact['dti_Rng'] = pd.cut(df_Impact['dti'], bins=5,precision =0,labels=['0-6','6-12','12-18','18-24','24-30'])
df_Impact.head(50)
```

# Conclusions

By running univariate Analysis on the 5627 Charged off loans.
The chances of the Application defaulting is higher  if

- The loan is offered to the applicant for  __36__ months tenure
- The loans is  offered at an interest rate of __(13-17) %__ .
- The loan amount disbursed is in the range of  __5000 - 10000__ 
- The loans is offered to an Applicant in the  __31000 -58000__ Income group
- The monthly installments is between  __145-274__ 
- The loan is given to an applicant who is staying on __rent__
- The loan is taken by appliacnt to repay existing debts(__debt consolidation__)
- The loan is taken by an Applicant who have __2-10__  open credit lines in the borrower's credit file.
- The loan is taken by an Applicant who has __2-20__   credit lines(open/closed) in the borrower's credit file.
- The loan is taken by an Applicant who has an employement of __10+__ years
- The loan is sanctioned to an Applicant with a Dti ratio is between __12-18__
- The loan is given to an Applicant with a loan status as __Not Verified__
- The loan is given to an Applicant with Grade  __'B'__
- The loan is given to an Applicant with s Sub Grade OF '__B5'__ .
- The highest default is seen in __CA__ state

__Observations on variables that had no impact on the default pattern__
- Even though Applicant who has made  0 enquiries in last 6 months are defaulting on loans this is not a great measure to analyse the defaulting probability
- Even though Applicant who has   0 derogatory public records are defaulting on loans this is not a great measure to analyse the defaulting probability
- No conclusion can be generated from the month in which the loan has been granted because loan disbursements every year has a different default pattern for every month and there is no general trend followed
